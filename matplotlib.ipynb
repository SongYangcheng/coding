```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10, 6), dpi=150, facecolor='white')
#
ax1 = fig.add_subplot(2, 2, 1)
t = np.linspace(0, 2*np.pi, 1000)
xh = 16*np.sin(t) ** 3
yh = 13*np.cos(t) - 5 * np.cos(2*t) - 3*np.cos(3*t) - np.cos(4*t)
ax1.plot(xh, yh, lw=2.2, color='red')
ax1.set_aspect('equal', adjustable='box') #xy, adjustable='box'
ax1.set_title("Heart", fontsize=12, weight='bold') #weight
ax1.grid(True, linestyle='--', alpha=0.6)
#
for spine in ['top', 'right']:
    ax1.spines[spine].set_visible(False) #
    #ax1.spines[spine] set_visible(False)
#
ax2 = fig.add_subplot(2, 2, 2)
t = np.linspace(0, 12*np.pi, 1000)
x_b = np.sin(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12) ** 5)
y_b = np.cos(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12) ** 5)
ax2.plot(x_b, y_b, 'b-', lw=1.5)
ax2.set_aspect('equal', adjustable='box')
ax2.set_title("Butterfly", fontsize=12, weight='bold')
ax2.grid(True, linestyle='--', alpha=0.6) #grid
#, KPI
ax3 = fig.add_subplot(2, 1, 2) #21
months = np.arange(1, 13)
sales = 120 + 20 * np.sin(2*np.pi*months/12) + 10 * np.random.randint(-12, 12, size=12)
ma3 = np.convolve(sales, np.ones(3)/3, mode='same') #3, convolvemode='same'
#np.ones(3)/3 31/33
avg = sales.mean()
#
bars = ax3.bar(months, sales, color='lightblue', label='Monthly Sales')
ax3.plot(months, ma3, color='orange', lw=2, label='3-Month MA')
#axhline, yavg
ax3.axhline(avg, color='red', ls='--',alpha=0.8,  lw=1.5, label=f'Annual Avg{avg:.0f}') #
#
i_peak = np.argmax(sales) #argmax
ax3.annotate('Peak', xy=(months[i_peak], sales[i_peak]), 
             xytext=(months[i_peak]+1, sales[i_peak]+10),
             arrowprops=dict(arrowstyle='->', color='green', lw=1.5),
             fontsize=10)
ax3.set_xlim(0.5, 12.5) #x
ax3.set_xticks(months) #x
ax3.set_xlabel("Month", fontsize=11) #x
ax3.set_ylabel("sales(k)", fontsize=11) 
ax3.set_title("KPI: 3-Month Moving Average", fontsize=12, weight='bold')
ax3.legend(ncols=3, frameon=False, fontsize=9) #ncols, lengend
ax3.grid(True, linestyle='--', alpha=0.6)

#
for spine in ['top', 'right']:
    ax3.spines[spine].set_visible(False)

plt.tight_layout() #
# plt.svefig('kpi_report.pdf', bbox_inches='tight') #
plt.show()
```

```python
##11+3)
import numpy as np
import matplotlib.pyplot as plt
months = np.arange(1, 13)
sales = np.cumsum(np.random.randint(80, 170, size=12)) #cusum
share = [20, 25, 15, 10, 30] #
cats = ['A', 'B', 'C', 'D', 'E'] #
regions = ['North', 'South', 'East', 'West'] #
region_sales = [250, 300, 200, 150] #
corr = np.corrcoef(np.random.randn(4, 60)) #randn460corrcoef

fig = plt.figure(figsize=(12, 7), dpi=200, facecolor='white')
#height_ratioshspacewspace
gs = fig.add_gridspec(2, 3, height_ratios=[2, 1], hspace=0.4, wspace=0.3)
#   
ax_main = fig.add_subplot(gs[0, :])
ax_main.plot(months, sales, 'b-o', lw=2, markersize=6) #markersize
ax_main.set_title("Monthly Sales Trend", fontsize=14, weight='bold')
ax_main.set_xlabel("Month", fontsize=12)
ax_main.set_ylabel("Sales (k)", fontsize=12)
ax_main.set_xticks(months) #x
ax_main.grid(True, linestyle='--', alpha=0.6)
# ax_main.legend(ncols=3, frameon=False, fontsize=9) #
# plt.show()
#1 
ax_pie = fig.add_subplot(gs[1, 0])
#autopct startangle,plt.cm.Pastel1.colors, labels
ax_pie.pie(share, labels=cats, autopct='%1.0f%%', startangle=90, colors=plt.cm.Pastel1.colors)
ax_pie.set_title("Category Share")

#2 
ax_bar = fig.add_subplot(gs[1, 1])
ax_bar.bar(regions, region_sales, color='lightgreen', alpha=1)
ax_bar.set_title("Sales by Region")
ax_bar.grid(True, linestyle='--', alpha=0.6, axis='y') #y
#3 
ax_corr = fig.add_subplot(gs[1, 2])
#imshow,cmap,vminvmax
im = ax_corr.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1) 
ax_corr.set_title("Correlation Matrix")
 #colorbar fractionpad
fig.colorbar(im, ax=ax_corr, fraction=0.046, pad=0.04)
fig.suptitle('Dashboard Layout ViaGridspec', fontsize=16, weight='bold')
# fig.savefig('gs_dashboard.pdf', dpi=300, bbox_inches='tight') #
```

```python
#2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

t = np.linspace(0, 2*np.pi, 1000)
x_h = 16*np.sin(t) ** 3
y_h = 13*np.cos(t) - 5 * np.cos(2*t) - 3*np.cos(3*t) - np.cos(4*t)

fig = plt.figure(figsize=(12, 6), dpi=200, facecolor='white')
gs = fig.add_gridspec(2, 2, width_ratios=[2,1], wspace=0.3)

#
ax_left = fig.add_subplot(gs[:, 0]) #gs[:,0]
ax_left.plot(x_h, y_h, color='red', lw=2)
ax_left.set_aspect('equal', adjustable='box')
ax_left.set_title("Heart", fontsize=14, weight='bold')
ax_left.grid(True, linestyle='--', alpha=0.6)
for spine in ['top', 'right']: #
    ax_left.spines[spine].set_visible(False)
#
ax_top_right = fig.add_subplot(gs[0, 1]) #gs[0,1]
t = np.linspace(0, 12*np.pi, 1000)
x_b = np.sin(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12) ** 5)
y_b = np.cos(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12) ** 5)
ax_top_right.plot(x_b, y_b, 'b-', lw=1.5)
ax_top_right.set_aspect('equal', adjustable='box')
ax_top_right.set_title("Butterfly", fontsize=12, weight='bold')
ax_top_right.axis('equal')
#xy
ax_top_right.set_xlabel("x", fontsize=11)
ax_top_right.set_ylabel("y", fontsize=11)
ax_top_right.grid(True, linestyle='--', alpha=0.6)
for spine in ['top', 'right']:
    ax_top_right.spines[spine].set_visible(False)
#
ax_bottom_right = fig.add_subplot(gs[1, 1]) #gs[1,1]
r3 = np.cos(2 * t)
x3, y3 = r3 * np.cos(t), r3 * np.sin(t)
ax_bottom_right.plot(x3, y3, 'g-', lw=2)
ax_bottom_right.set_aspect('equal', adjustable='box')
ax_bottom_right.set_title("Rose", fontsize=12, weight='bold')
ax_bottom_right.axis('equal')
ax_bottom_right.grid(True, linestyle='--', alpha=0.6)
for spine in ['top', 'right']:
    ax_bottom_right.spines[spine].set_visible(False)
fig.suptitle("Gridspec: Left Large + Right Two Small", fontsize=16, weight='bold')

```

```python
#+KPI)
import numpy as np
import matplotlib.pyplot as plt
#
#"Times New Roman" 
#"SimHei"
plt.rcParams['font.family'] = ["Times New Roman", "SimHei"]
plt.rcParams['axes.unicode_minus'] = False      #

rng = np.random.default_rng(42) #, default_rng()

months = np.arange(1, 13)
rev = 100 + np.cumsum(rng.integers(-12, 20, size=12)) #
ma3 = np.convolve(rev, np.ones(3)/3, mode='same') #3
avg = rev.mean()
mom = np.empty_like(rev, dtype=float) #rev
mom[0] = np.nan #NaN
mom[1:] = (rev[1:] - rev[:-1]) / np.maximum(rev[:-1], 1) * 100 #, maximum
#Main Figure
fig, ax = plt.subplots(figsize=(10, 6), dpi=150, facecolor='white')

fig.subplots_adjust(left=0.08, right=0.94, top=0.68, bottom=0.12) #

ax.plot(months, rev, marker='o', color='blue', lw=2, label='Revenue')
ax.plot(months, ma3, color='orange', lw=2, label='3-Month MA')
#axhline
ax.axhline(avg, color='red', ls='--', lw=1.5, label=f'Annual Avg {avg:.0f}')
#set_xlimx
ax.set_xlim(0.5, 12.5)
#set_xticksx
ax.set_xticks(months)
ax.set_xlabel("Month", fontsize=11)
ax.set_ylabel("Revenue (k)", fontsize=11)
ax.grid(True, linestyle='--', alpha=0.6)
#legend, 3, frameon=False
ax.legend(frameon=False, ncols=3)
for s in ('top', 'right'):
    ax.spines[s].set_visible(False)

#plt.axes()6 
ax_bar = plt.axes([0.08, 0.75, 0.40, 0.18], facecolor='white') #[left, bottom, width, height]
#pad
ax_bar.set_title("", fontsize=10, pad=0)
last6 = np.arange(len(months)-6+1, len(months)+1)
mom6 = mom[-6:]
colors = ['#ff9999' if v < 0 else '#66b3ff' for v in mom6] #
ax_bar.bar(last6, mom6, color=colors, alpha=0.9, width=0.6)
#axhline0y
ax_bar.axhline(0, color='#888', lw=1, alpha=0.7) #axhline
ax_bar.grid(True, ls='--', alpha=0.6, axis='y')

ax_bar.set_xlim(last6[0] - 0.3, last6[-1] + 0.3) #x
ax_bar.set_xticks(last6)
for s in ('top', 'right'):
    #spines
    ax_bar.spines[s].set_visible(False)

#
#add_axes()[left, bottom, width, height]leftbottom widthheight
ax_lines = fig.add_axes([0.52, 0.75, 0.40, 0.18], facecolor="white", frameon=True)
ax_lines.set_title("", fontsize=10, pad=0)
rev6 = rev[-6:]
#lw markersize
ax_lines.plot(last6, rev6, 'r-o', lw=1.8, markersize=6)
#grid
ax_lines.grid(True, ls='--', alpha=0.6)
ax_lines.set_xlim(last6[0] - 0.5, last6[-1] + 0.5)
ax_lines.set_xticks(last6)
for s in ('top', 'right'):
    ax_lines.spines[s].set_visible(False)
fig.suptitle("Main Figure with KPI Cards", fontsize=14, weight='bold')
# fig.savefig('kpi_cards.pdf', bbox_inches='tight') #
plt.show()
```

```python
# 
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(40)
months = np.arange(1, 13)
rev = 100 + np.cumsum(rng.integers(-12, 22, size=12)) #
#y(max/min)
ymin, ymax = rev.min(), rev.max()
m_ju1, m_aug = 7, 8 #
v_ju1, v_aug = rev[m_ju1-1], rev[m_aug-1] #
#7/1 -> 7/31 + 
days = 31
frac = np.arange(days)/days #31
x_daily = m_ju1 + frac #7.0 -> 7.967)
y_base = v_ju1 * (1 - frac) + v_aug * frac #, 
y_noise = rng.normal(scale=4.0, size=days) #
y_daily = y_base + y_noise #=+
#clip
y_daily = np.clip(y_daily, ymin, ymax) #y
#==========
fig, ax = plt.subplots(figsize=(12, 7), dpi=150, facecolor='white')
fig.subplots_adjust(left=0.08, right=0.94, top=0.62, bottom=0.10) #
#
ax.plot(months, rev, marker='o', color='blue', lw=2, label='Revenue')
ax.set_xlim(0.5, 12.5)
ax.set_xticks(months)
ax.set_ylim(ymin, ymax)
ax.set_xlabel("Month", fontsize=11)
ax.set_ylabel("Revenue (k)", fontsize=11)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(frameon=False) #
for s in ('top', 'right'):
    ax.spines[s].set_visible(False)

#===================
#left=0.08, right=0.94, 0.86 0.24
ax_top = plt.axes([0.08, 0.72, 0.86, 0.24], facecolor="white", frameon=True)
bar_w = 0.85 * (1/days)
print(y_daily)
ax_top.bar(x_daily, y_daily, width=bar_w, edgecolor='black', color='red', alpha=1, label="1->31")
ax_top.set_xlim(m_ju1 - bar_w/2, m_aug + bar_w/2)
ax_top.grid(True, ls='--', alpha=0.5)
for s in ('top', 'right'):
    ax_top.spines[s].set_visible(False)
ax_top.set_ylabel('', fontsize=10)
#78
y0, y1 = min(v_ju1, v_aug), max(v_ju1, v_aug) #y0 y1
pad = max((y1 - y0) * 0.20, 4.0) #4.0
#
#y0 - padlow 
low, high = min(y0 - pad, ymin), max(y1 + pad, ymax)
ax_top.set_ylim(low, high)

#===============
#indicate_inset_zoom
ax.indicate_inset_zoom(ax_top, edgecolor='blue', lw=1.0, alpha=0.95)
#======x========
#17-01 7-02........7-31
ticks_idx = np.arange(0, days,5)
ticks_pos = m_ju1 + ticks_idx/days #
#{d:02d}d0
tick_labels = [f'07-{d:02d}' for d in (ticks_idx + 1)]
#07-31
if ticks_idx[-1] != 30:
    ticks_pos = np.append(ticks_pos, m_ju1+30/days)
    tick_labels.append('07-31')
ax_top.set_xticks(ticks_pos)
#set_xticklabelsx, rotation
ax_top.set_xticklabels(tick_labels, fontsize=10, rotation=45) 

fig.suptitle(' + ', fontsize=14, weight='bold', y=1) #y=1
# fig.savefig()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = ["Times New Roman", "SimHei"]
plt.rcParams['axes.unicode_minus'] = False      #
years = np.arange(2013, 2020)
sales = [6.3, 7.1, 8.5, 9.0, 10.2, 11.5, 12.3] #
with plt.style.context('seaborn-v0_8-dark'): #seaborn-whitegrid
    fig, ax = plt.subplots(figsize=(8, 5), dpi=150, facecolor='white')
    ax.plot(years, sales, 'o-', color='blue', lw=2, markersize=6)
    ax.set_title("Annual Sales Trend", fontsize=14, weight='bold')
    ax.set_xlabel("Year", fontsize=12)
    ax.set_ylabel("Sales (k)", fontsize=12)
    ax.set_xticks(years)
    ax.grid(True, linestyle='--', alpha=0.6)
    for s in ('top', 'right'):
        ax.spines[s].set_visible(False)
    #
    for x, y in zip(years, sales):
        ax.text(x, y + 0.3, f'{y:.1f}', ha='center', va='bottom', fontsize=10)
    # fig.savefig('annual_sales_trend.pdf', bbox_inches='tight') #
    plt.show()
```
