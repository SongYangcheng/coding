```python
import matplotlib.pyplot as plt
import numpy as np
t = np.linspace(0, 2*np.pi, 1000)
#创建包含两行两列的子图网格
#同时设置画布大小和边框颜色，分辨率和背景颜色
fig,axs = plt.subplots(
    nrows=2, #子图行为2
    ncols=2, #子图列为2
    figsize=(10, 4), #画布大小
    dpi=200, #分辨率
    facecolor='None', #背景颜色
    frameon=True,  #开启画布边框
    linewidth=2, #边框线宽2
    edgecolor='red', #边框颜色为红色
    sharex=False, #共享X轴刻度
    sharey=False,
    width_ratios=[1,1], #第一列宽度为第三列的三分之一倍
    height_ratios=[1,1], #第二行高度为第一行的两倍
    subplot_kw={'facecolor': None}, #背景设置为透明
    gridspec_kw={'hspace':0.3, 'wspace':0.3} #子图间距为0.3
)
#在四个子图中绘制不同的内容
#左上心形线其中axs[0,0]表示第一行第一列
x1 = 16*np.sin(t) ** 3
y1 = 13*np.cos(t) - 5 * np.cos(2*t) - 3*np.cos(3*t) - np.cos(4*t)
axs[0,0].plot(x1, y1, 'r-', lw=2)
axs[0,0].set_title("Heart")
axs[0,0].axis('equal')
#绘制右上蝴蝶线axs[0,1]表示第一行第二列
x2 = np.sin(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12) ** 5)
y2 = np.cos(t) * (np.exp(np.cos(t)) - 2*np.cos(4*t) - np.sin(t/12) ** 5)
axs[0,1].plot(x2, y2, 'b-', lw=1.5)
axs[0,1].set_title("Butterfly")
axs[0,1].axis('equal')
#绘制左下玫瑰线axs[1,0]表示第二行第一列
r3 = np.cos(2 * t)
x3, y3 = r3 * np.cos(t), r3 * np.sin(t)
axs[1,0].plot(x3, y3, 'g-', lw=2)
axs[1,0].set_title("Rose")
axs[1,0].axis('equal')
#右下绘制阿基米德螺线axs[1,1]表示第二行第二列
theta = np.linspace(0, 10*np.pi, 1000)
r4 = theta
x4, y4 = r4 * np.cos(theta), r4 * np.sin(theta)
axs[1,1].plot(x4, y4, 'm-', lw=1.5)
axs[1,1].set_title("Archimedean Spiral")
axs[1,1].axis('equal')

#美化
for ax in axs.flat:
    ax.grid(True, linestyle='--', alpha=0.6) #添加网格线
    ax.label_outer() #仅显示外侧子图的坐标轴标签

# plt.savefig('subplot_demo.pdf', bbox_inches='tight')数据保存
plt.show()
```

### Pyplot接口和面向对象接口
#### Pyplot接口
是基于状态的接口由matplotlib.pyplot模块提供, 在当前活动的Figure图形和Axes子图上进行绘图，后台为你<font color="red">隐式的创建和管理Figure和Axes对象。</font>  

Pyplot接口函数通常以plt.开头，适合快速绘图和交互式使用包括以下常用函数：
- plt.plot(x, y, 'o-')  # 绘制线段图
- plt.scatter(x, y)     # 绘制散点图
- plt.bar(x, height)    # 绘制柱状图
- plt.hist(data, bins)  # 绘制直方图
- plt.pie(sizes, labels) # 绘制饼图
- plt.title('Title')    # 设置标题
- plt.xlabel('X-axis')  # 设置X轴标签
- plt.ylabel('Y-axis')  # 设置Y轴标签
- plt.grid(True)        # 显示网格
- plt.legend()          # 显示图例
- plt.xlim(min, max)    # 设置X轴范围
- plt.ylim(min, max)    # 设置Y轴范围
- plt.savefig('filename.png') # 保存图像
- plt.show()            # 显示图像
##### plt.subplot(nrows, ncols, index) 创建子图  
本质是Pyplot接口函数，但是也可也以看作是面向对象接口的一部分，允许在同一画布上创建多个子图，并返回当前激活的子图对象。  

其中的参数为:
- nrows: 子图行数
- ncols: 子图列数
- index: 当前子图位置，从1开始编号，按行优先
- plt.subplot(2, 3, 4) # 创建一个2行3列的子图，当前激活第4个子图

#### 面向对象接口
该方法更为灵活和强大，适合复杂绘图和定制化需求。通过<font color="red">显示的创建和管理Figure画布和Axes子图对象</font> ，可以更精细地控制图形的各个方面。
流程通常包括以下步骤：
1. 创建Figure对象：fig = plt.figure()
2. 添加Axes对象：ax = fig.add_subplot(111) 或 ax = fig.add_axes([left, bottom, width, height])
3. 使用Axes对象的方法进行绘图和定制
4. 显示或保存图像：plt.show() 或 fig.savefig('filename.png')  

面向对象接口通过创建Figure和Axes对象，适合复杂绘图和定制化需求包括以下常用方法：
- fig = plt.figure()    # 创建画布对象
- ax = fig.add_subplot(111) # 创建子图对象
- ax.plot(x, y, 'o-')  # 绘制线段图
- ax.scatter(x, y)     # 绘制散点图
- ax.bar(x, height)    # 绘制柱状图
- ax.hist(data, bins)  # 绘制直方图
- ax.pie(sizes, labels) # 绘制饼图
- ax.set_title('Title')    # 设置标题
- ax.set_xlabel('X-axis')  # 设置X轴标签
- ax.set_ylabel('Y-axis')  # 设置Y轴标签
- ax.grid(True)        # 显示网格
- ax.legend()          # 显示图例
- ax.set_xlim(min, max)    # 设置X轴范围
- ax.set_ylim(min, max)    # 设置Y轴范围
- fig.savefig('filename.png') # 保存图像
- plt.show()            # 显示图像

### plt.subplot和fig.subplots的区别
- plt.subplot()是Pyplot接口函数，用于在同一画布上创建多个子图，并返回当前激活的子图对象
- fig.subplots()是面向对象接口的方法，用于在已有画布上创建多个子图，并返回子图对象数组
- ax = plt.subplot(2, 3, 4) # 创建一个2行3列的子图，当前激活第4个子图，ax是当前子图对象
- axs = fig.subplots(2, 3) # 在已有画布fig上创建2行3列的子图，axs是一个2x3的数组，包含6个子图对象

### plt.subplots()和fig.subplots()的区别
- plt.subplots()是pyplot接口，用于创建画布和多个子图，返回画布对象和子图对象数组
- fig.subplots()是面向对象接口，用于在已有画布上创建多个子图，返回子图对象数组
- fig, axs = plt.subplots(2, 2) # 创建一个2行2列的子图，axs是一个2x2的数组，包含4个子图对象
- axs = fig.subplots(2, 2) # 在已有画布fig上创建2行2列的子图，axs是一个2x2的数组，包含4个子图对象

##### fig.add_subplot(nrows, ncols, index, **kwargs) 创建子图
- fig: 画布对象

- ncols: 子图列数
- index: 当前子图位置，从1开始编号，按行优先

### 进阶布局 --GridSpec
使用GridSpec和GridSpecFromSubplotSpec可以实现更复杂的子图布局和嵌套布局。  
fig.add_gridspec(nrows, ncols, **kwargs) 创建GridSpec布局
- nrows: 子图行数
- ncols: 子图列数
- width_ratios/height_ratios: 子图宽高比例
- wspace/hspace: 子图间距
- left/right/top/bottom: 子图边距
- gs = fig.add_gridspec(2, 2, wspace=0.4, hspace=0.4) # 创建一个2行2列的GridSpec布局，子图间距为0.4  


```python
##模板1：上1大图+下3小图（仪表盘常用)
import numpy as np
import matplotlib.pyplot as plt
months = np.arange(1, 13)
sales = np.cumsum(np.random.randint(80, 170, size=12)) #cusum计算累计和
share = [20, 25, 15, 10, 30] #市场份额数据
cats = ['A', 'B', 'C', 'D', 'E'] #类别标签
regions = ['North', 'South', 'East', 'West'] #地区标签
region_sales = [250, 300, 200, 150] #各地区销售数据
corr = np.corrcoef(np.random.randn(4, 60)) #randn生成4行60列的随机正态分布，corrcoef计算相关系数矩阵

fig = plt.figure(figsize=(12, 7), dpi=200, facecolor='white')
#height_ratios表示行高比例，hspace行间距，wspace列间距
gs = fig.add_gridspec(2, 3, height_ratios=[2, 1], hspace=0.4, wspace=0.3)
#大图   
ax_main = fig.add_subplot(gs[0, :])
ax_main.plot(months, sales, 'b-o', lw=2, markersize=6) #markersize表示点的大小
ax_main.set_title("Monthly Sales Trend", fontsize=14, weight='bold')
ax_main.set_xlabel("Month", fontsize=12)
ax_main.set_ylabel("Sales (k)", fontsize=12)
ax_main.set_xticks(months) #设置x轴刻度
ax_main.grid(True, linestyle='--', alpha=0.6)
# ax_main.legend(ncols=3, frameon=False, fontsize=9) #显示图例
# plt.show()
#子图1 饼图
ax_pie = fig.add_subplot(gs[1, 0])
#autopct显示百分比， startangle起始角度,plt.cm.Pastel1.colors使用预设的颜色, labels显示类别标签
ax_pie.pie(share, labels=cats, autopct='%1.0f%%', startangle=90, colors=plt.cm.Pastel1.colors)
ax_pie.set_title("Category Share")

#子图2 条形图
ax_bar = fig.add_subplot(gs[1, 1])
ax_bar.bar(regions, region_sales, color='lightgreen', alpha=1)
ax_bar.set_title("Sales by Region")
ax_bar.grid(True, linestyle='--', alpha=0.6, axis='y') #仅y轴显示网格即仅生成横向网格线
#子图3 热力图只有在数据呈现相关性的时候才有意义
ax_corr = fig.add_subplot(gs[1, 2])
#imshow显示矩阵,cmap颜色映射,vmin和vmax设置颜色映射范围
im = ax_corr.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1) 
ax_corr.set_title("Correlation Matrix")
 #colorbar为最右边的线条， fraction颜色条宽度占比，pad颜色条与图像间距
fig.colorbar(im, ax=ax_corr, fraction=0.046, pad=0.04)
fig.suptitle('Dashboard Layout ViaGridspec', fontsize=16, weight='bold')
# fig.savefig('gs_dashboard.pdf', dpi=300, bbox_inches='tight') #数据保存
```

```python
#主图+右上角KPI小卡片（绝对定位叠加)
import numpy as np
import matplotlib.pyplot as plt
#全局参数设置
#"Times New Roman" 当为英文时显示的字体
#"SimHei"当使用中文时显示黑体，除此之外还支持微软雅黑，宋体等
plt.rcParams['font.family'] = ["Times New Roman", "SimHei"]
plt.rcParams['axes.unicode_minus'] = False      #解决坐标负号显示异常

rng = np.random.default_rng(42) #创建随机数生成器, default_rng()生成均匀分布，类似于种子

months = np.arange(1, 13)
rev = 100 + np.cumsum(rng.integers(-12, 20, size=12)) #累计和
ma3 = np.convolve(rev, np.ones(3)/3, mode='same') #3个月移动平均
avg = rev.mean()
mom = np.empty_like(rev, dtype=float) #创建与rev形状相同的空数组
mom[0] = np.nan #第一个月环比增长率设为NaN
mom[1:] = (rev[1:] - rev[:-1]) / np.maximum(rev[:-1], 1) * 100 #计算环比增长率, maximum为避免除以零
#Main Figure
fig, ax = plt.subplots(figsize=(10, 6), dpi=150, facecolor='white')

fig.subplots_adjust(left=0.08, right=0.94, top=0.68, bottom=0.12) #调整子图位置

ax.plot(months, rev, marker='o', color='blue', lw=2, label='Revenue')
ax.plot(months, ma3, color='orange', lw=2, label='3-Month MA')
#axhline添加水平线
ax.axhline(avg, color='red', ls='--', lw=1.5, label=f'Annual Avg {avg:.0f}')
#set_xlim设置x轴范围
ax.set_xlim(0.5, 12.5)
#set_xticks设置x轴刻度
ax.set_xticks(months)
ax.set_xlabel("Month", fontsize=11)
ax.set_ylabel("Revenue (k)", fontsize=11)
ax.grid(True, linestyle='--', alpha=0.6)
#legend添加图例靠右边, 3列显示, frameon=False去掉图例边框
ax.legend(frameon=False, ncols=3)
for s in ('top', 'right'):
    ax.spines[s].set_visible(False)

#顶部左边用plt.axes()方近6期环比， 迷你柱状图
ax_bar = plt.axes([0.08, 0.75, 0.40, 0.18], facecolor='white') #[left, bottom, width, height]左下角坐标和宽高
#pad表示标题与图像间距
ax_bar.set_title("近六期环比", fontsize=10, pad=0)
last6 = np.arange(len(months)-6+1, len(months)+1)
mom6 = mom[-6:]
colors = ['#ff9999' if v < 0 else '#66b3ff' for v in mom6] #根据环比正负设置颜色
ax_bar.bar(last6, mom6, color=colors, alpha=0.9, width=0.6)
#axhline添加水平线其中0表示y轴位置
ax_bar.axhline(0, color='#888', lw=1, alpha=0.7) #axhline添加水平线
ax_bar.grid(True, ls='--', alpha=0.6, axis='y')

ax_bar.set_xlim(last6[0] - 0.3, last6[-1] + 0.3) #设置x轴范围
ax_bar.set_xticks(last6)
for s in ('top', 'right'):
    #spines表示边框
    ax_bar.spines[s].set_visible(False)

#顶部右边折线图
#add_axes()绝对定位[left, bottom, width, height]left和bottom表示左下角坐标， width和height表示宽高
ax_lines = fig.add_axes([0.52, 0.75, 0.40, 0.18], facecolor="white", frameon=True)
ax_lines.set_title("近六期环比趋势", fontsize=10, pad=0)
rev6 = rev[-6:]
#lw线宽， markersize点大小
ax_lines.plot(last6, rev6, 'r-o', lw=1.8, markersize=6)
#grid添加网格
ax_lines.grid(True, ls='--', alpha=0.6)
ax_lines.set_xlim(last6[0] - 0.5, last6[-1] + 0.5)
ax_lines.set_xticks(last6)
for s in ('top', 'right'):
    ax_lines.spines[s].set_visible(False)
fig.suptitle("Main Figure with KPI Cards", fontsize=14, weight='bold')
# fig.savefig('kpi_cards.pdf', bbox_inches='tight') #数据保存
plt.show()
```

```python
# 局部放大
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(40)
months = np.arange(1, 13)
rev = 100 + np.cumsum(rng.integers(-12, 22, size=12)) #累计和模拟月度收入
#主图y轴上下限(用月度值max/min来约束数据)
ymin, ymax = rev.min(), rev.max()
m_ju1, m_aug = 7, 8 #局部放大月份
v_ju1, v_aug = rev[m_ju1-1], rev[m_aug-1] #局部放大月份的值
#生成7/1 -> 7/31日期的日度轴 + 日数值
days = 31
frac = np.arange(days)/days #表示七月一和七月31之间的进度
x_daily = m_ju1 + frac #转化为月份坐标（7.0 -> 7.967)
y_base = v_ju1 * (1 - frac) + v_aug * frac #线性插值作为基准, 日度值的线性差值
y_noise = rng.normal(scale=4.0, size=days) #添加一些噪声
y_daily = y_base + y_noise #日度值=基准+噪声
#clip裁剪数据
y_daily = np.clip(y_daily, ymin, ymax) #裁剪到主图的y轴范围内
#=====画布与主图=====
fig, ax = plt.subplots(figsize=(12, 7), dpi=150, facecolor='white')
fig.subplots_adjust(left=0.08, right=0.94, top=0.62, bottom=0.10) #调整子图位置
#绘制主图
ax.plot(months, rev, marker='o', color='blue', lw=2, label='Revenue')
ax.set_xlim(0.5, 12.5)
ax.set_xticks(months)
ax.set_ylim(ymin, ymax)
ax.set_xlabel("Month", fontsize=11)
ax.set_ylabel("Revenue (k)", fontsize=11)
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(frameon=False) #不开启图例边框
for s in ('top', 'right'):
    ax.spines[s].set_visible(False)

#=========上方等宽小图==========
#与主图的左右编剧一致left=0.08, right=0.94, 宽取0.86， 上方高取0.24
ax_top = plt.axes([0.08, 0.72, 0.86, 0.24], facecolor="white", frameon=True)
bar_w = 0.85 * (1/days)
print(y_daily)
ax_top.bar(x_daily, y_daily, width=bar_w, edgecolor='black', color='red', alpha=1, label="日度1->31")
ax_top.set_xlim(m_ju1 - bar_w/2, m_aug + bar_w/2)
ax_top.grid(True, ls='--', alpha=0.5)
for s in ('top', 'right'):
    ax_top.spines[s].set_visible(False)
ax_top.set_ylabel('月度', fontsize=10)
#让主图标出的高亮矩形高度只围住7到8月的月度
y0, y1 = min(v_ju1, v_aug), max(v_ju1, v_aug) #y0下限， y1为上限
pad = max((y1 - y0) * 0.20, 4.0) #保证至少为4.0避免缓冲过小
#同时确保不超出主图的整体范围
#y0 - pad：数据下限减去缓冲的得到候选的low值， 
low, high = min(y0 - pad, ymin), max(y1 + pad, ymax)
ax_top.set_ylim(low, high)

#======自动标识主图的对应区间并连接到上图当中=========
#indicate_inset_zoom为主图添加放大区间
ax.indicate_inset_zoom(ax_top, edgecolor='blue', lw=1.0, alpha=0.95)
#======顶部小图的x轴标签========
#方案1：日期标签：7-01， 7-02........7-31
ticks_idx = np.arange(0, days,5)
ticks_pos = m_ju1 + ticks_idx/days #转换月小数位置
#{d:02d}表示整数d至少两位，不足前面补0
tick_labels = [f'07-{d:02d}' for d in (ticks_idx + 1)]
#确保包含07-31
if ticks_idx[-1] != 30:
    ticks_pos = np.append(ticks_pos, m_ju1+30/days)
    tick_labels.append('07-31')
ax_top.set_xticks(ticks_pos)
#set_xticklabels设置x轴刻度标签, rotation旋转角度
ax_top.set_xticklabels(tick_labels, fontsize=10, rotation=45) 

fig.suptitle('上方日度直方图 + 下方月度折线图', fontsize=14, weight='bold', y=1) #y=1表示标题位置
# fig.savefig()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = ["Times New Roman", "SimHei"]
plt.rcParams['axes.unicode_minus'] = False      #解决坐标负号显示
years = np.arange(2013, 2020)
sales = [6.3, 7.1, 8.5, 9.0, 10.2, 11.5, 12.3] #年度销售数据
with plt.style.context('seaborn-v0_8-dark'): #使用seaborn-whitegrid风格
    fig, ax = plt.subplots(figsize=(8, 5), dpi=150, facecolor='white')
    ax.plot(years, sales, 'o-', color='blue', lw=2, markersize=6)
    ax.set_title("Annual Sales Trend", fontsize=14, weight='bold')
    ax.set_xlabel("Year", fontsize=12)
    ax.set_ylabel("Sales (k)", fontsize=12)
    ax.set_xticks(years)
    ax.grid(True, linestyle='--', alpha=0.6)
    for s in ('top', 'right'):
        ax.spines[s].set_visible(False)
    #添加数据标签
    for x, y in zip(years, sales):
        ax.text(x, y + 0.3, f'{y:.1f}', ha='center', va='bottom', fontsize=10)
    # fig.savefig('annual_sales_trend.pdf', bbox_inches='tight') #数据保存
    plt.show()
```
