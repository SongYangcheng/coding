#### set_context()函数
set_context()函数用于设置Seaborn图形的上下文环境，以调整图形的整体外观和比例。它允许用户根据不同的展示需求（如论文、演示等）来调整图形的字体大小、线条粗细等参数，从而提高图形的可读性和美观性。

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize #标准化颜色映射
import pandas as pd
from matplotlib.cm import ScalarMappable #颜色映射,生成颜色条
plt.rcParams['font.sans-serif'] = ['SimHei']  #设置中文字体为SimHei
# tips = pd.read_csv(r'D:\python_demo\coding\data\seaborndata\tips.csv')
tips = sns.load_dataset('tips') #加载内置数据集seaborn自带数据
tips.columns = ['总账单','小费','性别','是否吸烟','星期',
                '用餐时间段','人数']
tips['总额'] = tips['总账单'] + tips['小费']

tips['小费率'] = (tips['小费'] / tips['总账单']).round(4)


top5 = tips.sort_values('小费率',ascending = False).head()
tips['总账单_jitter'] = tips['总账单'] + np.random.normal(0, 0.5, size=len(tips))
fig, axes = plt.subplots(2, 3, figsize=(12, 10))
ax = axes[0, 0]
print(tips.head())
sns.scatterplot(data=tips, x='总账单', y='小费', ax=ax, color='purple', s=120)
ax.set_title('散点图示例')
ax.set_xlabel('消费金额 ($)')
ax.set_ylabel('小费金额 ($)')

#子图B
ax = axes[0, 1]
sns.scatterplot(data=tips, x='总额', y='总账单', hue='性别', palette='Set1', ax=ax)
ax.set_title('B.颜色编码：总额 vs 总账单')
ax.set_xlabel('总额')
ax.set_ylabel('总账单')
ax.legend(title='性别',
          frameon=True,
          loc='upper left',
          fontsize=10,
          title_fontsize=10)
#子图C
ax = axes[0, 2]
sns.scatterplot(data=tips, x='总额',
                 y='小费率', hue='性别',
                 style='是否吸烟',size='人数',
                 sizes=(20, 160),#将人数映射到点面积范围最小20，最大160
                 alpha=0.7, #点透明度
                 linewidth=0.6, #点的边缘线宽为0.6
                 hue_order=['Male', 'Female'], #强制颜色排序，Male再图例中排前
                 style_order=['Yes', 'No'], #强制样式排序，Yes在前
                 s=99,
                 palette='Set1', ax=ax)

ax.set_title('C.颜色编码：总额 vs 小费率')
ax.set_xlabel('总额')
ax.set_ylabel('小费率')
ax.legend(title='性别',
          frameon=True,
          loc='upper left',
          fontsize=10,
          title_fontsize=10,
          borderaxespad=1)
#子图D
ax = axes[1, 0]
sns.scatterplot(data=tips, x='总额',
                 y='小费', hue='星期',
                 style='是否吸烟',size='小费率',
                 palette='viridis', #使用viridis调色板（蓝绿黄色渐变）
                 alpha=0.7, #点透明度
                 linewidth=0.6, #点的边缘线宽为0.6
                 hue_order=['Thur', 'Fri', 'Sat', 'Sun'], #强制颜色排序，Male再图例中排前
                 style_order=['Yes', 'No'], #强制样式排序，Yes在前
                 s=99,
                 size_norm=Normalize(
                     vmin=tips['小费率'].min(),
                     vmax=tips['小费率'].max()), #自定义大小映射范围
                     ax=ax)

ax.set_title('D.颜色编码：总额 vs 小费率')
ax.set_xlabel('总额')
ax.set_ylabel('小费')
ax.legend(title='星期/小费率',
          frameon=True,
          loc='upper left',
          fontsize=10,
          title_fontsize=10)
#子图E
ax = axes[1, 1]
sns.scatterplot(
    data=tips,
    x='总账单_jitter',
    y='小费',
    hue='用餐时间段',
    style='是否吸烟',
    palette='coolwarm', #使用coolwarm调色板
    size='人数', #大小按照人数映射
    sizes=(30, 160),
    alpha=0.7,
    linewidth=0.6,
    edgecolor='white',
    ax=ax
)
ax.set_title('E.抖动散点图：总账单 vs 小费')
ax.set_xlabel('总账单 (抖动)')
ax.set_ylabel('小费')
ax.legend(title='用餐时段/是否吸烟/人数',
          frameon=True,
          loc='upper left',
          fontsize=9,
          title_fontsize=10)
#子图F
ax = axes[1, 2]
sns.scatterplot(
    data=tips,
    x='总账单_jitter',
    y='小费',
    hue='用餐时间段',
    style='是否吸烟',
    palette='coolwarm', #使用coolwarm调色板
    size='人数', #大小按照人数映射
    sizes=(30, 160),
    alpha=0.7,
    linewidth=0.6,
    edgecolor='white',
    ax=ax,
#指定调色板
)
#坐标与网格
ax.set_xlim(0, 60)
ax.set_ylim(0, 12)
ax.grid(True, linestyle='--', alpha=0.3)
sns.despine(ax=ax) #移除顶部和右侧坐标轴
ax.set_title('F.自定义坐标与网格：总账单 vs 小费')
ax.set_xlabel('总账单 (抖动)')
ax.set_ylabel('小费')
for _, r in top5.iterrows(): #遍历top5数据行
    ax.annotate(
        f"{r['性别']}, {r['小费率']*100:.1f}%",
        (r['总账单'], r['小费']),
        xytext=(5, -5), #偏移位置文本标签 (Text) 实际放置的位置, 箭头指向的数据点位置
        textcoords='offset points', #偏移单位
        fontsize=9,
        color='black',
        weight='bold',
        arrowprops=dict(facecolor='black', shrink=0.05) #箭头属性
    )
ax.legend(title='用餐时段/是否吸烟/人数',
          frameon=True,
          loc='upper left',
          fontsize=9,
          title_fontsize=10) #图例设置
fig.suptitle('Seaborn.scatterplot 联系：3X2画布示例', fontsize=16,
             y=1.02 #主标题大小为16，y=1.02使其略高于子图顶部
)
# for ax in axes.flat: #遍历所有子图轴对象
#     ax.set_xlim(0, 60) #统一x轴范围
#     ax.set_ylim(0, 15) #统一y轴范围
plt.tight_layout()
plt.show()
```

#### relplot()函数和scatterplot()函数的区别
relplot()函数是Seaborn库中的一个高级绘图函数，用于创建关系图（relational plots）。它可以用于绘制散点图（scatter plots）和线图（line plots），并且支持分面（faceting）功能，可以根据一个或多个分类变量将数据分成多个子图。
scatterplot()函数是Seaborn库中的一个基础绘图函数，专门用于创建散点图。它不支持分面功能，主要用于绘制单个散点图。

#### cut参数
cut参数用于控制核密度估计的范围。它指定了在数据范围之外绘制密度估计的距离。默认情况下，cut参数的值为3，这意味着密度估计将在数据范围之外延伸3个标准差。如果将cut参数设置为0，则密度估计将仅在数据范围内绘制。

### 条形图-barplot()函数
barplot()函数用于创建条形图，显示分类变量与数值变量之间的关系。它通过计算每个类别的均值（或其他聚合统计量）来绘制条形图，并可以显示误差条以表示数据的变异性。

#### 参数说明
统计参数  
- estimator：用于计算每个x值对应的y值的函数。默认情况下，estimator=None，表示使用所有数据点进行估计。可以将其设置为numpy.mean、numpy.median等函数，以计算每个x值对应的y值的平均值、中位数等。
- errorbar：用于指定误差条的类型。可以设置为'sd'（标准差）、'ci'（置信区间）等。默认情况下，errorbar=None，表示不绘制误差条。
- capsize：用于指定误差条的帽子大小。默认情况下，capsize=None，表示不绘制帽子。可以将其设置为一个正数，以指定帽子的大小。
- order：用于指定数据点的绘制顺序。默认情况下，order=None，表示按照数据的顺序绘制。可以将其设置为一个整数列表，以指定数据点的绘制顺序。  
- hue: 用于指定分组变量。可以将数据按照hue变量的不同取值进行分组，并为每个组绘制不同的颜色。默认情况下，hue=None，表示不进行分组。
样式参数
- palette：用于指定颜色调色板。可以设置为预定义的调色板名称（如'deep'、'muted'、'bright'等），也可以设置为自定义的颜色列表。默认情况下，palette=None，表示使用默认的颜色调色板。
- color：用于指定单一颜色。可以设置为颜色名称（如'red'、'blue'等）、十六进制颜色代码（如'#FF0000'）等。默认情况下，color=None，表示使用默认颜色。  
布局参数
- dodge：用于指定是否对分组数据进行错开显示。默认情况下，dodge=False，表示不进行错开显示。可以将其设置为True，以对分组数据进行错开显示。  
- legend：用于指定是否显示图例。默认情况下，legend='auto'，表示根据需要显示图例。可以将其设置为True或False，以强制显示或隐藏图例。
- ax：用于指定绘图的坐标轴对象。默认情况下，ax=None，表示在当前坐标轴上绘图。可以将其设置为一个matplotlib的Axes对象，以在指定的坐标轴上绘图。

### 箱线图绘制-boxplot()函数
箱线图（Box Plot）是一种用于显示数据分布情况的统计图表。它通过五个统计量来描述数据的分布：最小值、第一四分位数（Q1）、中位数（Q2）、第三四分位数（Q3）和最大值。箱线图还可以显示异常值（outliers），即远离其他数据点的值。   
上边缘：Q3+1.5IQR  
上四分位数(Q3)  
中位数：Q2  
下四分位数(Q1)  
下边缘：Q1-1.5IQR  
IQR(四分位距)=Q3-Q1
##### boxplot()函数参数
- x, y, hue：用于指定分类变量和数值变量。x和y分别表示分类变量和数值变量，hue用于指定分组变量。
- order, hue_order：用于指定分类变量和分组变量的绘制顺序。
- palette：用于指定颜色调色板。
- width：用于指定箱线图的宽度。
- showfliers：用于指定是否显示异常值, 离群点类似气泡圆圈状在图形中。
- showmeans：用于指定是否显示均值。
- notch: 用于指定是否绘制缺口箱线图。
- orient：用于指定箱线图的方向。
- whis: 用于指定须的范围。
- whiskerprops: 用于自定义须的样式属性。
- boxprops应箱体, flierprops异常值, medianprops中位数, meanprops均值, capprops帽子, whiskerprops须的样式：用于自定义箱线图各部分的样式属性。分别对应箱体、异常值、中位数、均值、帽子和须的样式属性。
- fill：用于指定是否填充箱体颜色。
- dodge：用于指定是否对分组数据进行错开显示。 
- palette：用于指定颜色调色板。可选的值包括预定义的调色板名称（如'deep'、'muted'、'bright'等）或自定义的颜色列表。     
常用技巧：  
boxplot + swarmplot:不重叠数据点展示分布， 箱型图和蜂群图结合 
boxplot + stripplot:展示数据分布和具体数据点， 箱型图和抖动散点图结合
##### swarmplot()函数和stripplot()函数区别
swarmplot()函数用于创建蜂群图（swarm plot），它通过调整数据点的位置来避免重叠，从而更清晰地展示数据的分布情况。swarmplot()函数适用于较小的数据集，因为它会根据数据点的数量自动调整位置，以确保每个数据点都能被清晰地看到。
stripplot()函数用于创建抖动散点图（strip plot），它通过在数据点周围添加随机抖动来避免重叠，从而展示数据的分布情况。stripplot()函数适用于较大的数据集，因为它不会根据数据点的数量调整位置，而是通过添加抖动来减少重叠。

### 小提琴图-violinplot()函数
小提琴图（Violin Plot）是一种结合了箱线图和核密度估计图的统计图表。它通过显示数据的分布情况和概率密度来提供更丰富的信息。小提琴图的形状类似于小提琴，因此得名。
#### violinplot()函数参数
- x, y, hue：用于指定分类变量和数值变量。x和y分别表示分类变量和数值变量，hue用于指定分组变量。
- order, hue_order：用于指定分类变量和分组变量的绘制顺序
- palette：用于指定颜色调色板。
- width：用于指定小提琴图的宽度。
- inner：用于指定小提琴图内部的显示方式。可选的值包括'box'（显示箱线图）、'quartile'（显示四分位数）、'point'（显示数据点）和'stick'（显示数据点的垂直线）。
- cut：用于指定核密度估计的范围。默认情况下，cut=1，会在数据范围之外延伸1个标准差。可以将其设置为0，以仅在数据范围内绘制密度估计。
- scale：用于指定小提琴图的缩放方式。可选的值包括'area'（根据面积缩放）、'count'（根据计数缩放）和'width'（根据宽度缩放）。
- bw：用于指定核密度估计的带宽。带宽控制了密度估计的平滑程度。较小的带宽会导致更尖锐的密度估计，而较大的带宽会导致更平滑的密度估计。
- bw_adjust：用于调整核密度估计的带宽。它是一个乘数，用于调整默认带宽的大小。较大的bw_adjust值会增加带宽，从而使密度估计更平滑；较小的bw_adjust值会减小带宽，从而使密度估计更尖锐。
- inner：用于指定小提琴图内部的显示方式。可选的值包括'box'（显示箱线图）、'quartile'（显示四分位数）、'point'（显示数据点）和'stick'（显示数据点的垂直线）。
- density_estimator：用于指定核密度估计的方法。可选的值包括'scipy'（使用SciPy库进行密度估计）和'statsmodels'（使用Statsmodels库进行密度估计）。
- density_norm：用于指定密度估计的归一化方式。可选的值包括'area'（根据面积归一化）、'count'（根据计数归一化）和'probability'（根据概率归一化）。

#### 常用组合
boxplot + violinplot: 展示数据分布的多样性， 箱型图和小提琴图结合
violinplot + swarmplot: 展示数据分布和具体数据点， 小提琴图和蜂群图结合
violinplot + stripplot: 展示数据分布和具体数据点， 小提琴图和抖动散点图结合

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 加载多个数据集
penguins = sns.load_dataset('penguins').dropna()
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')

print(f"Penguins数据集形状: {penguins.shape}")
print(f"Tips数据集形状: {tips.shape}")
print(f"Iris数据集形状: {iris.shape}")
print(tips.columns.tolist())
print(iris.columns.tolist())
# ============================================================
# 画布 1: 基础用法和核心参数
# ============================================================
fig1, axes1 = plt.subplots(2, 5, figsize=(24, 10))  # 创建2行5列子图布局，总画布大小24x10英寸
fig1.suptitle('Seaborn Violinplot 教程 - 画布1: 基础用法与核心参数',  # 设置整个画布的总标题
              fontsize=16,  # 总标题字体大小16
              fontweight='bold',  # 总标题加粗显示
              y=0.995)  # 总标题垂直位置略低于顶部（避免裁剪）

# ========== 子图 A：最基础的小提琴图 ==========
ax = axes1[0, 0]  # 获取第1行第1列子图对象
sns.violinplot(data=penguins,  # 数据源：企鹅数据集
               x='species',  # x轴：企鹅物种（分类变量）
               y='body_mass_g',  # y轴：体重（数值变量）
               ax=ax,  # 指定绘制到当前子图
               color='steelblue')  # 设置单一颜色为钢蓝色
ax.set_title('A. 基础小提琴图\n密度+分布', fontsize=11, fontweight='bold')  # 子图标题说明密度分布功能
ax.set_xlabel('企鹅种类', fontsize=10)  # x轴标签：企鹅种类
ax.set_ylabel('体重 (g)', fontsize=10)  # y轴标签：体重（克）
ax.text(0.02,  # 文本水平位置：左边距2%
        0.98,  # 文本垂直位置：顶部98%
        '小提琴图特点:\n显示密度分布\n对称的核密度估计\n宽度=数据密度',  # 详细说明小提琴图核心特点
        transform=ax.transAxes,  # 使用相对坐标系
        fontsize=8,  # 字体大小8
        verticalalignment='top',  # 顶部对齐
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))  # 圆角浅黄色半透明背景框

# ========== 子图 B：添加hue分组 ==========
ax = axes1[0, 1]  # 获取第1行第2列子图对象
sns.violinplot(data=penguins,  # 数据源
               x='species',  # x轴：物种
               y='body_mass_g',  # y轴：体重
               hue='sex',  # hue参数：按性别分组，并排显示
               ax=ax)  # 绘制到当前子图，自动配色
ax.set_title('B. 分组小提琴图\nhue分组', fontsize=11, fontweight='bold')  # 标题说明分组功能
ax.set_xlabel('企鹅种类', fontsize=10)  # x轴标签
ax.set_ylabel('体重 (g)', fontsize=10)  # y轴标签
ax.legend(title='性别', loc='upper left', fontsize=8)  # 图例：标题"性别"，左上角，字体8
ax.text(0.02,  # 文本位置
        0.98,
        '参数: hue="sex"\n并排显示不同组',  # 说明hue参数并排显示效果
        transform=ax.transAxes,
        fontsize=9,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))  # 浅蓝色背景框

# ========== 子图 C：split分割显示 ==========
ax = axes1[0, 2]  # 获取第1行第3列子图对象
sns.violinplot(data=penguins,  # 数据源
               x='species',  # x轴：物种
               y='body_mass_g',  # y轴：体重
               hue='sex',  # hue：按性别分组
               split=True,  # split=True：左右分割显示（仅限2组hue）
               ax=ax)  # 绘制到当前子图，自动配色
ax.set_title('C. 分割小提琴图\nsplit=True', fontsize=11, fontweight='bold')  # 标题说明分割功能
ax.set_xlabel('企鹅种类', fontsize=10)  # x轴标签
ax.set_ylabel('体重 (g)', fontsize=10)  # y轴标签
ax.legend(title='性别', loc='upper left', fontsize=8)  # 图例设置
ax.text(0.02,  # 文本位置
        0.98,
        'split=True:\n左右分割显示\n更紧凑的对比\n仅适用于2组hue',  # 说明split适用条件和优势
        transform=ax.transAxes,
        fontsize=8,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.5))  # 浅黄色背景框

# ========== 子图 D：横向小提琴图 ==========
ax = axes1[0, 3]  # 获取第1行第4列子图对象
sns.violinplot(
               data=tips,  # 数据源：tips小费数据集
               x='total_bill',  # x轴：账单总额（数值）
               y='day',  # y轴：星期（分类）-交换实现横向
               hue='day',  # hue：按星期着色
               palette='viridis',  # viridis连续色彩调色板
               legend=False,  # 隐藏图例
               ax=ax)  # 绘制到当前子图
ax.set_title('D. 横向小提琴图\nTips数据集', fontsize=11, fontweight='bold')  # 标题说明横向和数据集
ax.set_xlabel('账单金额 ($)', fontsize=10)  # x轴标签：美元
ax.set_ylabel('星期', fontsize=10)  # y轴标签
ax.text(0.02,  # 文本位置
        0.98,
        '参数: x和y互换\n使用tips数据集\n展示消费分布',  # 说明横向参数和应用场景
        transform=ax.transAxes,
        fontsize=8,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.5))  # 浅绿色背景框

# ========== 子图 E：inner参数 - box ==========
ax = axes1[0, 4]  # 获取第1行第5列子图对象
sns.violinplot(data=penguins,  # 数据源
               x='species',  # x轴：物种
               y='flipper_length_mm',  # y轴：鳍长
               hue='species',  # 分组：按物种着色
               palette='Set2',  # Set2柔和色彩调色板
               legend=False,  # 隐藏图例
               inner='box',  # inner="box"：内部显示微型箱线图
               ax=ax)  # 绘制目标
ax.set_title('E. 内部显示箱线图\ninner="box"', fontsize=11, fontweight='bold')  # 标题说明内部箱线
ax.set_xlabel('企鹅种类', fontsize=10)  # x轴标签
ax.set_ylabel('鳍长 (mm)', fontsize=10)  # y轴标签
ax.text(0.02,  # 文本位置
        0.98,
        'inner="box":\n内部显示微型箱线图\n结合两种图优点',  # 说明箱线+密度组合优势
        transform=ax.transAxes,
        fontsize=8,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lavender', alpha=0.5))  # 淡紫色背景框

# ========== 子图 F：inner参数 - quartile ==========
ax = axes1[1, 0]  # 获取第2行第1列子图对象
sns.violinplot(data=tips,  # 数据源：tips数据集
               x='day',  # x轴：星期
               y='tip',  # y轴：小费
               hue='day',  # 分组：按星期
               palette='coolwarm',  # 冷暖色系调色板
               legend=False,  # 隐藏图例
               inner='quartile',  # inner="quartile"：显示Q1、Q2、Q3四分位线
               ax=ax)  # 绘制目标
ax.set_title('F. 显示四分位数\ninner="quartile"', fontsize=11, fontweight='bold')  # 标题说明四分位显示
ax.set_xlabel('星期', fontsize=10)  # x轴标签
ax.set_ylabel('小费 ($)', fontsize=10)  # y轴标签
ax.text(0.02,  # 文本位置
        0.98,
        'inner="quartile":\n显示四分位线\n标记Q1,Q2,Q3',  # 说明四分位线标记
        transform=ax.transAxes,
        fontsize=8,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='mistyrose', alpha=0.5))  # 玫瑰色背景框

# ========== 子图 G：inner参数 - point ==========
ax = axes1[1, 1]  # 获取第2行第2列子图对象
sns.violinplot(data=penguins,  # 数据源
               x='island',  # x轴：岛屿
               y='bill_length_mm',  # y轴：喙长
               hue='island',  # 分组：按岛屿
               palette='muted',  # 柔和色彩调色板
               legend=False,  # 隐藏图例
               inner='point',  # inner="point"：显示每个原始数据点
               ax=ax)  # 绘制目标
ax.set_title('G. 显示单个点\ninner="point"', fontsize=11, fontweight='bold')  # 标题说明数据点显示
ax.set_xlabel('岛屿', fontsize=10)  # x轴标签
ax.set_ylabel('喙长 (mm)', fontsize=10)  # y轴标签
ax.text(0.02,  # 文本位置
        0.98,
        'inner="point":\n显示每个数据点\n适合小数据集',  # 说明适用场景
        transform=ax.transAxes,
        fontsize=8,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightcyan', alpha=0.5))  # 浅青色背景框

# ========== 子图 H：inner参数 - stick ==========
ax = axes1[1, 2]  # 获取第2行第3列子图对象
sns.violinplot(data=iris,  # 数据源：iris鸢尾花数据集
               x='species',  # x轴：花种
               y='sepal_length',  # y轴：萼片长度
               hue='species',  # 分组：按花种
               palette='Spectral',  # 光谱色调色板
               legend=False,  # 隐藏图例
               inner='stick',  # inner="stick"：每条竖线代表1个数据点
               ax=ax)  # 绘制目标
ax.set_title('H. 显示棒状标记\ninner="stick" (Iris)', fontsize=11, fontweight='bold')  # 标题说明棒状标记
ax.set_xlabel('鸢尾花种类', fontsize=10)  # x轴标签
ax.set_ylabel('萼片长度 (cm)', fontsize=10)  # y轴标签
ax.text(0.02,  # 文本位置
        0.98,
        'inner="stick":\n显示竖线标记\n每条线=1个数据',  # 说明棒状标记含义
        transform=ax.transAxes,
        fontsize=8,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='peachpuff', alpha=0.5))  # 桃色背景框

# ========== 子图 I：inner参数 - None ==========
ax = axes1[1, 3]  # 获取第2行第4列子图对象
sns.violinplot(data=penguins,  # 数据源
               x='species',  # x轴：物种
               y='body_mass_g',  # y轴：体重
               hue='species',  # 分组
               palette='tab10',  # 10色分类调色板
               legend=False,  # 隐藏图例
               inner=None,  # inner=None：不显示任何内部标记
               ax=ax)  # 绘制目标
ax.set_title('I. 无内部标记\ninner=None', fontsize=11, fontweight='bold')  # 标题说明纯密度显示
ax.set_xlabel('企鹅种类', fontsize=10)  # x轴标签
ax.set_ylabel('体重 (g)', fontsize=10)  # y轴标签
ax.text(0.02,  # 文本位置
        0.98,
        'inner=None:\n只显示密度曲线\n最简洁的形式',  # 说明最简洁形式
        transform=ax.transAxes,
        fontsize=9,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))  # 浅黄色背景框

# ========== 子图 J：调整带宽 bw_adjust ==========
ax = axes1[1, 4]  # 获取第2行第5列子图对象
sns.violinplot(data=tips,  # 数据源：tips数据集
               x='time',  # x轴：时间段（午餐/晚餐）
               y='total_bill',  # y轴：账单总额
               hue='time',  # 分组：按时间段
               palette='rocket',  # rocket红黑渐变调色板
               legend=False,  # 隐藏图例
               bw_adjust=0.5,  # bw_adjust=0.5：带宽调整为默认的50%（更细节）
               ax=ax)  # 绘制目标
ax.set_title('J. 调整平滑度\nbw_adjust=0.5', fontsize=11, fontweight='bold')  # 标题说明平滑度调整
ax.set_xlabel('时间段', fontsize=10)  # x轴标签
ax.set_ylabel('账单金额 ($)', fontsize=10)  # y轴标签
ax.text(0.02,  # 文本位置
        0.98,
        'bw_adjust:\n<1: 更多细节\n>1: 更平滑\n默认1.0',  # 详细说明带宽调整效果
        transform=ax.transAxes,
        fontsize=8,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.5))  # 浅蓝色背景框

plt.tight_layout()  # 自动调整子图间距，避免重叠
plt.show()  # 显示画布1的10个子图
```

### 热力图-heatmap()函数和clustermap()函数
heatmap()函数用于创建热力图，显示二维数据的数值大小。它通过颜色的深浅来表示数值的大小，适用于展示矩阵数据或相关性矩阵等。
clustermap()函数用于创建聚类热力图，结合了热力图和层次聚类分析。它通过对数据进行聚类分析，将相似的数据点分组，并在热力图中显示这些分组的关系。

#### melt()函数和pivot()函数
这两个函数是数据处理和变换中非常重要的工具。是相反的操作。
melt()函数用于将数据从宽格式转换为长格式。它通过将多个列合并为两个列（变量名和变量值）来实现这一转换，适用于需要进行分组或聚合分析的数据处理。
pivot()函数用于将数据从长格式转换为宽格式。它通过将一个列的唯一值转换为多个列来实现这一转换，适用于需要重新组织数据以便进行分析或可视化的场景。
