```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize #标准化颜色映射
import pandas as pd
from matplotlib.cm import ScalarMappable #颜色映射,生成颜色条
plt.rcParams['font.sans-serif'] = ['SimHei']  #设置中文字体为SimHei
# tips = pd.read_csv(r'D:\python_demo\coding\data\seaborndata\tips.csv')
tips = sns.load_dataset('tips') #加载内置数据集seaborn自带数据
tips.columns = ['总账单','小费','性别','是否吸烟','星期',
                '用餐时间段','人数']
tips['总额'] = tips['总账单'] + tips['小费']

tips['小费率'] = (tips['小费'] / tips['总账单']).round(4)


top5 = tips.sort_values('小费率',ascending = False).head()
tips['总账单_jitter'] = tips['总账单'] + np.random.normal(0, 0.5, size=len(tips))
fig, axes = plt.subplots(2, 3, figsize=(12, 10))
ax = axes[0, 0]
print(tips.head())
sns.scatterplot(data=tips, x='总账单', y='小费', ax=ax, color='purple', s=120)
ax.set_title('散点图示例')
ax.set_xlabel('消费金额 ($)')
ax.set_ylabel('小费金额 ($)')

#子图B
ax = axes[0, 1]
sns.scatterplot(data=tips, x='总额', y='总账单', hue='性别', palette='Set1', ax=ax)
ax.set_title('B.颜色编码：总额 vs 总账单')
ax.set_xlabel('总额')
ax.set_ylabel('总账单')
ax.legend(title='性别',
          frameon=True,
          loc='upper left',
          fontsize=10,
          title_fontsize=10)
#子图C
ax = axes[0, 2]
sns.scatterplot(data=tips, x='总额',
                 y='小费率', hue='性别',
                 style='是否吸烟',size='人数',
                 sizes=(20, 160),#将人数映射到点面积范围最小20，最大160
                 alpha=0.7, #点透明度
                 linewidth=0.6, #点的边缘线宽为0.6
                 hue_order=['Male', 'Female'], #强制颜色排序，Male再图例中排前
                 style_order=['Yes', 'No'], #强制样式排序，Yes在前
                 s=99,
                 palette='Set1', ax=ax)

ax.set_title('C.颜色编码：总额 vs 小费率')
ax.set_xlabel('总额')
ax.set_ylabel('小费率')
ax.legend(title='性别',
          frameon=True,
          loc='upper left',
          fontsize=10,
          title_fontsize=10,
          borderaxespad=1)
#子图D
ax = axes[1, 0]
sns.scatterplot(data=tips, x='总额',
                 y='小费', hue='星期',
                 style='是否吸烟',size='小费率',
                 palette='viridis', #使用viridis调色板（蓝绿黄色渐变）
                 alpha=0.7, #点透明度
                 linewidth=0.6, #点的边缘线宽为0.6
                 hue_order=['Thur', 'Fri', 'Sat', 'Sun'], #强制颜色排序，Male再图例中排前
                 style_order=['Yes', 'No'], #强制样式排序，Yes在前
                 s=99,
                 size_norm=Normalize(
                     vmin=tips['小费率'].min(),
                     vmax=tips['小费率'].max()), #自定义大小映射范围
                     ax=ax)

ax.set_title('D.颜色编码：总额 vs 小费率')
ax.set_xlabel('总额')
ax.set_ylabel('小费')
ax.legend(title='星期/小费率',
          frameon=True,
          loc='upper left',
          fontsize=10,
          title_fontsize=10)
#子图E
ax = axes[1, 1]
sns.scatterplot(
    data=tips,
    x='总账单_jitter',
    y='小费',
    hue='用餐时间段',
    style='是否吸烟',
    palette='coolwarm', #使用coolwarm调色板
    size='人数', #大小按照人数映射
    sizes=(30, 160),
    alpha=0.7,
    linewidth=0.6,
    edgecolor='white',
    ax=ax
)
ax.set_title('E.抖动散点图：总账单 vs 小费')
ax.set_xlabel('总账单 (抖动)')
ax.set_ylabel('小费')
ax.legend(title='用餐时段/是否吸烟/人数',
          frameon=True,
          loc='upper left',
          fontsize=9,
          title_fontsize=10)
#子图F
ax = axes[1, 2]
sns.scatterplot(
    data=tips,
    x='总账单_jitter',
    y='小费',
    hue='用餐时间段',
    style='是否吸烟',
    palette='coolwarm', #使用coolwarm调色板
    size='人数', #大小按照人数映射
    sizes=(30, 160),
    alpha=0.7,
    linewidth=0.6,
    edgecolor='white',
    ax=ax,
#指定调色板
)
#坐标与网格
ax.set_xlim(0, 60)
ax.set_ylim(0, 12)
ax.grid(True, linestyle='--', alpha=0.3)
sns.despine(ax=ax) #移除顶部和右侧坐标轴
ax.set_title('F.自定义坐标与网格：总账单 vs 小费')
ax.set_xlabel('总账单 (抖动)')
ax.set_ylabel('小费')
for _, r in top5.iterrows(): #遍历top5数据行
    ax.annotate(
        f"{r['性别']}, {r['小费率']*100:.1f}%",
        (r['总账单'], r['小费']),
        xytext=(5, -5), #偏移位置文本标签 (Text) 实际放置的位置
        textcoords='offset points', #偏移单位
        fontsize=9,
        color='black',
        weight='bold',
    )
ax.legend(title='用餐时段/是否吸烟/人数',
          frameon=True,
          loc='upper left',
          fontsize=9,
          title_fontsize=10) #图例设置
fig.suptitle('Seaborn.scatterplot 联系：3X2画布示例', fontsize=16,
             y=1.02 #主标题大小为16，y=1.02使其略高于子图顶部
)
# for ax in axes.flat: #遍历所有子图轴对象
#     ax.set_xlim(0, 60) #统一x轴范围
#     ax.set_ylim(0, 15) #统一y轴范围
plt.tight_layout()
plt.show()
```
