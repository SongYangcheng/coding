```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotx

# 读取Excel文件
df = pd.read_excel('./data/scatter.xlsx')
# 使用marplotx库中的浅色风格
with plt.style.context(matplotx.styles.pitaya_smoothie['light']):
    # 创建图形对象，设置画布大小为12x6英寸
    plt.figure(figsize=(12, 6), dpi=200)

    # 获取当前图形的坐标轴对象（Axes），用于后续设置坐标轴属性
    ax = plt.gca()
    # 设置绘图区域（坐标轴内）的背景色为浅灰色
    ax.set_facecolor('#f0f0f0')   # 浅灰色背景

    # 绘制系列的散点图
    # x轴数据为df中的'x1'列，y轴数据为'Series1'列
    plt.scatter(df['x1'], df['Series1'],
                edgecolors='black',   # 设置数据点边缘颜色为黑色
                linewidth=0.4,        # 设置数据点边缘线宽0.4
                alpha=0.8,            # 设置数据点透明度为0.8（0为完全透明，1为完全不透明）
                label='Series1',      # 设置该系列的标签：用于图例显示
                marker='P',           # 设置数据点形状为菱形（'P'代表pentagon，五边形/菱形）
                s=80)                 # 设置数据点大小为80（数值越大点越大）

    # 绘制第二个系列的散点图
    # x轴数据为df的'x2'列，y轴数据为'Series2'列
    plt.scatter(df['x2'], df['Series2'],
                edgecolors='black',   # 数据点边缘颜色为黑色
                linewidth=0.4,        # 边缘线宽0.4
                alpha=0.8,            # 透明度0.8
                label='Series2',      # 系列标签，用于图例
                marker='s',           # 数据点形状为正方形（'s'代表square）
                s=80)                 # 设置数据点大小为80（数值越大点越大）

    # 绘制第三个系列的散点图
    # x轴数据为df的'x3'列，y轴数据为'Series3'列
    plt.scatter(df['x3'], df['Series3'],
                edgecolors='black',   # 数据点边缘颜色为黑色
                linewidth=0.4,        # 边缘线宽0.4
                alpha=0.8,            # 透明度0.8
                label='Series3',      # 系列标签，用于图例
                marker='o',           # 数据点形状为圆形（'o'代表circle）
                s=80)                 # 设置数据点大小为80（数值越大点越大）

    # 为图形添加标题
    plt.title('The Multi-series Scatterplot', fontsize=16)
    # 设置y轴标签为'Value'，字体大小为14
    plt.ylabel('Value', fontsize=14)
    # 设置x轴标签
    plt.xlabel('This is x axis', fontsize=14)
    # 显示图例，设置为左上角
    plt.legend(loc=2)

    # 为图形添加网格线
    plt.grid(alpha=0.8,            # 网格线透明度
            linestyle='--',        # 网格线样式虚线
            linewidth=0.8,         # 网格线线宽
            color='white')         # 网格线颜色为白色

    # 保存图形
    plt.savefig('./outputs/scatter_plot.pdf',
                dpi=300,              # 分辨率
                bbox_inches='tight',  # 自动调整边框，紧凑型布局
                pad_inches=0.0)       # 边框与图形内容的间距为0
    
    # 保存图形
    plt.savefig('./outputs/scatter_plot.png',
                dpi=300,              # 分辨率
                bbox_inches='tight',  # 自动调整边框，紧凑型布局
                pad_inches=0.0)       # 边框与图形内容的间距为0

    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotx

# 读取CSV文件
mtcars = pd.read_csv('./data/mtcars.csv')
# 查看数据前2行内容
mtcars.head(2)

# 从数据中提取x轴数据
x = mtcars['wt']       # 重量
# 从数据中提取y轴数据
y = mtcars['mpg']      # 燃油效率
# 定义点的大小数据
size = mtcars['disp']  # 排量（用于点的大小）
# 定义点的颜色数据
fill = mtcars['disp']  # 排量（用于点的颜色填充）

# 使用matplotx的浅色主题
with plt.style.context(matplotx.styles.pitaya_smoothie['light']):
    # 创建图形对象
    plt.figure(figsize=(12, 6), dpi=300)

    # 绘制散点图
    scatter = plt.scatter(
        x=x,                    # x轴数据为汽车重量
        y=y,                    # y轴数据为燃油效率
        c=fill,                 # 点的颜色由排量数据决定（fill变量）
        s=size/3,               # 点的大小由排量数据决定，除以3是为了避免点过大
        linewidth=0.5,          # 点的边缘线宽度为0.5
        edgecolor="k",          # 点的边缘颜色设置为黑色（"k"是黑色缩写）
        cmap='RdYlBu_r'         # 使用RdYlBu_r颜色映射（红-黄-蓝反向渐变）
    )

    # 为散点图添加颜色条，用于说明颜色对应的排量数值，关联之前的scatter对象
    cbar = plt.colorbar(scatter)
    # 设置颜色条的标签文本为 'Displacement (disp)'
    cbar.set_label('Displacement (disp)', fontsize=12)  # 颜色条标签（排量）

    # 设置x轴的标签文本
    plt.xlabel('Weight', fontsize=12)         # x轴标签（重量）
    # 设置y轴的标签文本
    plt.ylabel('Miles per Gallon (MPG)', fontsize=12)  # y轴标签（燃油效率）

    # 设置标题
    plt.title('Scatter Plot of Weight vs MPG (Colored By Displacement)', fontsize=16)

    # 保存图形
    plt.savefig(
        './output_image/sca_02.pdf',
        dpi=300,             
        bbox_inches='tight',        # 自动调整图片边框，确保所有标签完整显示
        pad_inches=0.1              # 图片边缘与内容的间距为0.1英寸，避免内容贴边
    )

    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotx
import matplotlib as mpl

month_df = pd.DataFrame({
    'month': np.arange(1, 6),
    'count': [20, 2276, 5232, 5565, 4943]
})

#计算关键统计值
total = month_df['count'].sum()
peak_idx = month_df['count'].idxmax()
peak_month = month_df.loc[peak_idx, 'month']
peak_value = month_df.loc[peak_idx, 'count']
peak_ratio = peak_value / total * 100

#颜色配置
base_cmap = mpl.colormaps.get_cmap('Blues') #选择Blues颜色映射
colors = [base_cmap(i/(len(month_df)-1)) for i in range(len(month_df))]
colors[peak_idx] = '#FF5733' #高亮峰值颜色
fig, ax = plt.subplots(figsize=(7, 7), dpi=200)
#绘制饼图
wedges, texts, autotexts = ax.pie(
    month_df['count'],
    labels=[f'月份 {m}' for m in month_df['month']],
    autopct='%1.1f%%', #显示百分比, 自动格式化为1位小数
    startangle=140, #起始角度
    colors=colors,
    counterclock=False, #顺时针方向
    pctdistance=0.85, #百分比标签距离圆心的距离
    labeldistance=1.1, #标签距离圆心的距离
    textprops={'fontsize': 12, 'color': 'black'}
)
centre_circle = plt.Circle((0, 0), 0.70, fc='white') #添加白色圆圈以创建环形图
ax.add_artist(centre_circle)
```
