```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotx

# Excel
df = pd.read_excel('./data/scatter.xlsx')
# marplotx
with plt.style.context(matplotx.styles.pitaya_smoothie['light']):
    # 12x6
    plt.figure(figsize=(12, 6), dpi=200)

    # Axes
    ax = plt.gca()
    # 
    ax.set_facecolor('#f0f0f0')   # 

    # 
    # xdf'x1'y'Series1'
    plt.scatter(df['x1'], df['Series1'],
                edgecolors='black',   # 
                linewidth=0.4,        # 0.4
                alpha=0.8,            # 0.801
                label='Series1',      # 
                marker='P',           # 'P'pentagon/
                s=80)                 # 80

    # 
    # xdf'x2'y'Series2'
    plt.scatter(df['x2'], df['Series2'],
                edgecolors='black',   # 
                linewidth=0.4,        # 0.4
                alpha=0.8,            # 0.8
                label='Series2',      # 
                marker='s',           # 's'square
                s=80)                 # 80

    # 
    # xdf'x3'y'Series3'
    plt.scatter(df['x3'], df['Series3'],
                edgecolors='black',   # 
                linewidth=0.4,        # 0.4
                alpha=0.8,            # 0.8
                label='Series3',      # 
                marker='o',           # 'o'circle
                s=80)                 # 80

    # 
    plt.title('The Multi-series Scatterplot', fontsize=16)
    # y'Value'14
    plt.ylabel('Value', fontsize=14)
    # x
    plt.xlabel('This is x axis', fontsize=14)
    # 
    plt.legend(loc=2)

    # 
    plt.grid(alpha=0.8,            # 
            linestyle='--',        # 
            linewidth=0.8,         # 
            color='white')         # 

    # 
    plt.savefig('./outputs/scatter_plot.pdf',
                dpi=300,              # 
                bbox_inches='tight',  # 
                pad_inches=0.0)       # 0
    
    # 
    plt.savefig('./outputs/scatter_plot.png',
                dpi=300,              # 
                bbox_inches='tight',  # 
                pad_inches=0.0)       # 0

    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotx

# CSV
mtcars = pd.read_csv('./data/mtcars.csv')
# 2
mtcars.head(2)

# x
x = mtcars['wt']       # 
# y
y = mtcars['mpg']      # 
# 
size = mtcars['disp']  # 
# 
fill = mtcars['disp']  # 

# matplotx
with plt.style.context(matplotx.styles.pitaya_smoothie['light']):
    # 
    plt.figure(figsize=(12, 6), dpi=300)

    # 
    scatter = plt.scatter(
        x=x,                    # x
        y=y,                    # y
        c=fill,                 # fill
        s=size/3,               # 3
        linewidth=0.5,          # 0.5
        edgecolor="k",          # "k"
        cmap='RdYlBu_r'         # RdYlBu_r--
    )

    # scatter
    cbar = plt.colorbar(scatter)
    #  'Displacement (disp)'
    cbar.set_label('Displacement (disp)', fontsize=12)  # 

    # x
    plt.xlabel('Weight', fontsize=12)         # x
    # y
    plt.ylabel('Miles per Gallon (MPG)', fontsize=12)  # y

    # 
    plt.title('Scatter Plot of Weight vs MPG (Colored By Displacement)', fontsize=16)

    # 
    plt.savefig(
        './output_image/sca_02.pdf',
        dpi=300,             
        bbox_inches='tight',        # 
        pad_inches=0.1              # 0.1
    )

    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotx
import matplotlib as mpl

month_df = pd.DataFrame({
    'month': np.arange(1, 6),
    'count': [20, 2276, 5232, 5565, 4943]
})

#
total = month_df['count'].sum()
peak_idx = month_df['count'].idxmax()
peak_month = month_df.loc[peak_idx, 'month']
peak_value = month_df.loc[peak_idx, 'count']
peak_ratio = peak_value / total * 100

#
base_cmap = mpl.colormaps.get_cmap('Blues') #Blues
colors = [base_cmap(i/(len(month_df)-1)) for i in range(len(month_df))]
colors[peak_idx] = '#FF5733' #
fig, ax = plt.subplots(figsize=(7, 7), dpi=200)
#
wedges, texts, autotexts = ax.pie(
    month_df['count'],
    labels=[f' {m}' for m in month_df['month']],
    autopct='%1.1f%%', #, 1
    startangle=140, #
    colors=colors,
    counterclock=False, #
    pctdistance=0.85, #
    labeldistance=1.1, #
    textprops={'fontsize': 12, 'color': 'black'}
)
centre_circle = plt.Circle((0, 0), 0.70, fc='white') #
ax.add_artist(centre_circle)
```
